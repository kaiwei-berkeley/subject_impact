```python
import pandas as pd
```

```python
df = pd.read_csv('year_trend.csv')
df.head()
```

```python
df.shape
```

```python
df.dtypes
```

```python
string = df.iloc[0,2]
string_split = string.split(',')
string_split
new_dict = {}
for j in range(len(string_split)):
    tmp = string_split[j].strip(" '{}").split(':')
    new_dict[int(tmp[0])] = int(tmp[1])
new_dict
```

```python
###### convert 'year_counts' column to dictionary ######
new_col = []
for i in range(len(df)):
    string = df.iloc[i,2]
    string_split = string.split(',')
    new_dict = {}
    for j in range(len(string_split)):
        tmp = string_split[j].strip(" '{}").split(':')
        new_dict[int(tmp[0])] = int(tmp[1])
    new_col.append(new_dict)
    print('Done with', i)
    
df['year_counts_norm'] = new_col
df.head()
```

```python
df.drop('years_counts', axis = 1, inplace = True)
df.head()
```

```python
for i in range(len(df)):
    tot = sum(df.iloc[i,2].values())
    for key in df.iloc[i,2]:
        df.iloc[i,2][key] = round(df.iloc[i,2][key] / tot, 2)
df.head()
```

```python
df.to_csv('year_trend_norm.csv', index = False)
```
