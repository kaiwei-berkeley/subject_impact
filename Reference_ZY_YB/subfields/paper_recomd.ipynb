```python
# read paper_subject_match
import pandas as pd
match = pd.read_csv('paper_subject_match_subfield.csv')
```

```python
###### convert 'references' column to list ######
new_col = []
for i in range(len(reference)):
    words = reference.loc[i,'refereneces']
    words_split = words.split(',')
    for j in range(len(words_split)):
        words_split[j] = words_split[j].strip(" '[]")
    new_col.append(words_split)
    
reference['references'] = new_col
reference.head()
```

```python
reference.drop('refereneces', axis = 1, inplace = True)
reference.drop('issn', axis = 1, inplace = True)
reference.drop('subjects', axis = 1, inplace = True)
```

```python
##### add paper's subject as new col ######
reference['paper_field'] = match['subfield']
reference.head()
```

```python
# reset 'id' column as index
match = match.set_index('id',)
reference = reference.set_index('id',)
```

```python
pairs = {key:value for key,value in zip(match.index,match['subfield'])}
```

```python
pairs[999] = 9999
```

```python
###### find fields of reference ######
reference_subjects = []
row = 1
for i in reference.index:
    references = reference.loc[i,'references']
    index = []
    for x in references:
        if x in reference.index:
            index.append(x)
        else:
            index.append(999)
    
    subjects = [pairs[key] for key in index]
    reference_subjects.append(subjects)
    if row % 10000 == 0:
        print('row ' + str(row) + ' finished')
    row +=1

reference['reference_fields'] = reference_subjects
```

```python
###### separate pairs ######
import numpy as np
impact = reference.reset_index().loc[:,['paper_field', 'reference_fields', 'references']]
def explode(df, lst_cols, fill_value='', preserve_index=False):
    # make sure `lst_cols` is list-alike
    if (lst_cols is not None
        and len(lst_cols) > 0
        and not isinstance(lst_cols, (list, tuple, np.ndarray, pd.Series))):
        lst_cols = [lst_cols]
    # all columns except `lst_cols`
    idx_cols = df.columns.difference(lst_cols)
    # calculate lengths of lists
    lens = df[lst_cols[0]].str.len()
    # preserve original index values    
    idx = np.repeat(df.index.values, lens)
    # create "exploded" DF
    res = (pd.DataFrame({
                col:np.repeat(df[col].values, lens)
                for col in idx_cols},
                index=idx)
             .assign(**{col:np.concatenate(df.loc[lens>0, col].values)
                            for col in lst_cols}))
    # append those rows that have empty lists
    if (lens == 0).any():
        # at least one list in cells is empty
        res = (res.append(df.loc[lens==0, idx_cols], sort=False))
    # revert the original index order
    res = res.sort_index()
    # reset index if requested
    if not preserve_index:        
        res = res.reset_index(drop=True)
    return res

impact2 = explode(impact, ['reference_fields', 'references'])
impact2.dropna(inplace = True)
```

```python
impact2.head()
```

```python
impact3 = impact2[impact2['reference_fields'] != 9999]
impact3.head()
```

```python
9999 in impact3['reference_fields'].unique()
```

```python
###### extract references for the same group ######
groups = impact3.groupby(['paper_field', 'reference_fields']).size().reset_index()
ref = []
for i in range(len(groups)):
    p = groups.iloc[i,0]
    r = groups.iloc[i,1]
    tmp = list(impact3[ (impact3['paper_field'] == p) & (impact3['reference_fields'] == r) ]['references'].value_counts()[:5].index) 
    ref.append(tmp)
    if i % 100 == 0:
        print('Done with', i)
    
groups['references'] = ref
```
