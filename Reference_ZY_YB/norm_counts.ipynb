```python
import pandas as pd
```

```python
counts = pd.read_csv('pairs_counts.csv')
counts
```

```python
# normalization
for i in counts.loc[counts['paper_subject'] == counts['reference_subject'],].index:
    counts.loc[i, 'counts'] = 0
counts
```

```python
tmp = counts.set_index('paper_subject',)
tmp
```

```python
norm = []
for i in range(len(tmp)):
    res = tmp.iloc[i,1] / sum(tmp.loc[tmp.index[i],'counts'])
    norm.append(res)
counts['norm_counts'] = norm
counts
```

```python
counts.drop('counts', axis = 1, inplace = True)
```

```python
counts.to_csv('norm_counts.csv', index = False)
```
